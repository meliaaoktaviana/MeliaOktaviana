```python

import matplotlib.pyplot as plt
# Parameter awal dari Kasus 1
h0 = 100  # ketinggian awal (m)
v0 = 0    # kecepatan awal (m/s)
g = 9.8   # percepatan gravitasi (m/s²)
dt = 1    # interval waktu (s)
t_max = 10  # waktu maksimum (s)
# Buat list waktu dari 0 hingga 10 detik dengan interval 1 detik
times = list(range(0, t_max + 1))
# Hitung posisi (ketinggian) h(t) = h0 - (1/2)*g*t²
# (mengambil arah ke bawah sebagai positif untuk kecepatan dan percepatan, sehingga posisi menurun)
posisi = [h0 - 0.5 * g * t**2 for t in times]
# Hitung kecepatan v(t) = g * t (kecepatan ke bawah, positif)
kecepatan = [g * t for t in times]
# Percepatan konstan a(t) = g
percepatan = [g for _ in times]
# Simpan hasil ke dalam tiga list terpisah (seperti diminta)
# times: [0,1,2,...,10]
# posisi: list posisi
# kecepatan: list kecepatan
# percepatan: list percepatan
# Tampilkan hasil dalam bentuk tabel sederhana untuk verifikasi
print("Waktu (s) | Posisi (m) | Kecepatan (m/s) | Percepatan (m/s²)")
for i in range(len(times)):
    print(f"{times[i]:8.0f} | {posisi[i]:10.2f} | {kecepatan[i]:15.2f} | {percepatan[i]:18.2f}")
# Tentukan nilai maksimum posisi serta kecepatan benda
max_posisi = max(posisi)  # Maks di t=0
max_kecepatan = max(kecepatan)  # Maks di t=10
print(f"\nNilai maksimum posisi: {max_posisi} m (pada t=0 s)")
print(f"Nilai maksimum kecepatan: {max_kecepatan} m/s (pada t=10 s)")
# Penjelasan fisika singkat
print("\nPenjelasan Hubungan Grafik dan Fenomena Fisika:")
print("- Grafik posisi vs waktu: Parabola menurun, menunjukkan percepatan konstan menyebabkan perpindahan kuadratik.")
print("- Grafik kecepatan vs waktu: Garis lurus naik, sesuai v = u + at (u=0, a=g).")
print("- Grafik percepatan vs waktu: Horizontal, gravitasi konstan.")
print("- Secara keseluruhan: Mengilustrasikan kinematika 1D dengan percepatan konstan (gerak jatuh bebas).")
# Buat dan tampilkan grafik ketiganya dalam satu plot
plt.figure(figsize=(10, 6))  # Atur ukuran figure
plt.plot(times, posisi, label='Posisi (ketinggian)', color='blue', marker='o', linewidth=2)  # Grafik posisi vs waktu
plt.plot(times, kecepatan, label='Kecepatan', color='red', marker='s', linewidth=2)  # Grafik kecepatan vs waktu
plt.plot(times, percepatan, label='Percepatan', color='green', marker='^', linewidth=2)  # Grafik percepatan vs waktu
plt.xlabel('Waktu (s)')  # Label sumbu x
plt.ylabel('Nilai')  # Label sumbu y (skala berbeda, tapi untuk ilustrasi)
plt.title('Grafik Posisi, Kecepatan, dan Percepatan vs Waktu - Gerak Jatuh Bebas')  # Judul
plt.legend()  # Legenda
plt.grid(True)  # Grid untuk kemudahan baca
plt.show()  # Tampilkan grafik
# Tambahan untuk Jupyter/Colab: Pastikan plot inline
%matplotlib inline
```

```python
# Kasus 2 – Jatuh Bebas dengan Kecepatan Awal
# Sebuah benda dilempar ke bawah dengan kecepatan awal 10 m/s dari ketinggian 120 meter,
# percepatan gravitasi g = 9,8 m/s2. Hitung posisi, kecepatan, dan percepatan setiap 0,5 detik
# selama 10 detik, simpan dalam list, tampilkan grafik posisi, kecepatan, dan percepatan
# dalam satu plot, dan tentukan nilai maksimum posisi serta kecepatan benda
# Import library yang diperlukan
import matplotlib.pyplot as plt
# Parameter awal
h0 = 120   # ketinggian awal (m)
v0 = 10    # kecepatan awal ke bawah (m/s)
g = 9.8    # percepatan gravitasi (m/s²)
dt = 0.5   # interval waktu (s)
t_max = 10 # waktu maksimum (s)
# Buat list waktu dari 0 hingga 10 detik dengan interval 0.5 detik
# Jumlah titik: (10 / 0.5) + 1 = 21 titik
times = [i * dt for i in range(0, int(t_max / dt) + 1)]
# Hitung posisi (ketinggian) h(t) = h0 - v0 * t - (1/2) * g * t²
# (arah ke bawah positif: posisi ketinggian menurun)
posisi = [h0 - v0 * t - 0.5 * g * t**2 for t in times]
# Hitung kecepatan v(t) = v0 + g * t (kecepatan ke bawah, positif)
kecepatan = [v0 + g * t for t in times]
# Percepatan konstan a(t) = g
percepatan = [g for _ in times]
# Simpan hasil ke dalam tiga list terpisah
# times: list waktu
# posisi: list posisi (ketinggian)
# kecepatan: list kecepatan
# percepatan: list percepatan
# Tampilkan hasil dalam bentuk tabel sederhana untuk verifikasi
print("Waktu (s) | Posisi (m) | Kecepatan (m/s) | Percepatan (m/s²)")
print("-" * 50)
for i in range(len(times)):
    print(f"{times[i]:8.1f} | {posisi[i]:10.2f} | {kecepatan[i]:15.2f} | {percepatan[i]:18.2f}")
# Tentukan nilai maksimum posisi serta kecepatan benda
max_posisi = max(posisi)  # Maks di t=0
max_kecepatan = max(kecepatan)  # Maks di t=10
print(f"\nNilai maksimum posisi: {max_posisi} m (pada t=0 s)")
print(f"Nilai maksimum kecepatan: {max_kecepatan} m/s (pada t=10 s)")
# Penjelasan singkat hubungan grafik dan fenomena fisika
print("\nPenjelasan Singkat:")
print("- Posisi vs waktu: Parabola menurun lebih cepat daripada kasus tanpa v0, karena kecepatan awal mempercepat jatuh.")
print("- Kecepatan vs waktu: Garis lurus naik dengan slope g, mulai dari 10 m/s.")
print("- Percepatan vs waktu: Horizontal konstan di g=9.8 m/s².")
print("- Waktu menyentuh tanah (h=0): Sekitar t = [-v0 + sqrt(v0² + 2 g h0)] / g ≈ 3.8 s (dihitung secara matematis).")
print("- Grafik mengilustrasikan kinematika dengan percepatan konstan dan kecepatan awal.")
# Buat dan tampilkan grafik posisi, kecepatan, dan percepatan dalam satu plot
plt.figure(figsize=(12, 7))  # Atur ukuran figure lebih besar karena lebih banyak titik
plt.plot(times, posisi, label='Posisi (ketinggian)', color='blue', marker='o', markersize=4, linewidth=2)
plt.plot(times, kecepatan, label='Kecepatan', color='red', marker='s', markersize=4, linewidth=2)
plt.plot(times, percepatan, label='Percepatan', color='green', marker='^', markersize=4, linewidth=2)
plt.xlabel('Waktu (s)')
plt.ylabel('Nilai')
plt.title('Grafik Posisi, Kecepatan, dan Percepatan vs Waktu - Jatuh Bebas dengan Kecepatan Awal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kasus 3 – Planet Lain (Gravitasi Berbeda)
# Bayangkan sebuah benda jatuh bebas di bulan dengan gravitasi g = 1,6 m/s2, dari
# ketinggian 50 meter, tanpa kecepatan awal. Hitung posisi, kecepatan, dan percepatan setiap
# 2 detik selama 20 detik, simpan dalam list, tampilkan grafik ketiganya dalam satu plot, dan
# tentukan nilai maksimum posisi serta kecepatan benda.
# Import library yang diperlukan
import matplotlib.pyplot as plt
# Parameter awal
h0 = 50    # ketinggian awal (m)
v0 = 0     # kecepatan awal (m/s)
g = 1.6    # percepatan gravitasi Bulan (m/s²)
dt = 2     # interval waktu (s)
t_max = 20 # waktu maksimum (s)
# Buat list waktu dari 0 hingga 20 detik dengan interval 2 detik
# Jumlah titik: (20 / 2) + 1 = 11 titik
times = list(range(0, t_max + 1, int(dt)))
# Hitung posisi (ketinggian) h(t) = h0 - (1/2) * g * t²
# (arah ke bawah positif: posisi ketinggian menurun)
posisi = [h0 - 0.5 * g * t**2 for t in times]
# Hitung kecepatan v(t) = g * t (kecepatan ke bawah, positif)
kecepatan = [g * t for t in times]
# Percepatan konstan a(t) = g
percepatan = [g for _ in times]
# Simpan hasil ke dalam tiga list terpisah
# times: list waktu
# posisi: list posisi (ketinggian)
# kecepatan: list kecepatan
# percepatan: list percepatan
# Tampilkan hasil dalam bentuk tabel sederhana untuk verifikasi
print("Waktu (s) | Posisi (m) | Kecepatan (m/s) | Percepatan (m/s²)")
print("-" * 50)
for i in range(len(times)):
    print(f"{times[i]:8.0f} | {posisi[i]:10.2f} | {kecepatan[i]:15.2f} | {percepatan[i]:18.2f}")
# Tentukan nilai maksimum posisi serta kecepatan benda
max_posisi = max(posisi)  # Maks di t=0
max_kecepatan = max(kecepatan)  # Maks di t=20
print(f"\nNilai maksimum posisi: {max_posisi} m (pada t=0 s)")
print(f"Nilai maksimum kecepatan: {max_kecepatan} m/s (pada t=20 s)")
# Penjelasan singkat hubungan grafik dan fenomena fisika
print("\nPenjelasan Singkat:")
print("- Posisi vs waktu: Parabola menurun lebih lambat daripada di Bumi (g=9.8), karena gravitasi Bulan lebih kecil.")
print("- Kecepatan vs waktu: Garis lurus naik dengan slope kecil (g=1.6), mulai dari 0 m/s.")
print("- Percepatan vs waktu: Horizontal konstan di g=1.6 m/s².")
print("- Waktu menyentuh tanah (h=0): t = sqrt(2 h0 / g) ≈ 7.9 s (lebih lama daripada di Bumi).")
print("- Grafik mengilustrasikan efek gravitasi rendah: jatuh lebih lambat, cocok untuk simulasi di planet lain.")
# Buat dan tampilkan grafik posisi, kecepatan, dan percepatan dalam satu plot
plt.figure(figsize=(12, 7))  # Atur ukuran figure
plt.plot(times, posisi, label='Posisi (ketinggian)', color='blue', marker='o', markersize=6, linewidth=2)
plt.plot(times, kecepatan, label='Kecepatan', color='red', marker='s', markersize=6, linewidth=2)
plt.plot(times, percepatan, label='Percepatan', color='green', marker='^', markersize=6, linewidth=2)
plt.xlabel('Waktu (s)')
plt.ylabel('Nilai')
plt.title('Grafik Posisi, Kecepatan, dan Percepatan vs Waktu - Jatuh Bebas di Bulan (g=1.6 m/s²)')
plt.legend()
plt.grid(True)
plt.show()
```
