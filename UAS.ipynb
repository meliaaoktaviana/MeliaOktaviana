```python
import matplotlib.pyplot as plt
import numpy as np
h = 100
g = 9.8
v0 = 10
t = list(range(11))
s = [h + v0 * ti + 0.5 * g * ti ** 2 for ti in t]
v = [v0 * ti for ti in t]
a = g * np.ones_like(t)
#plot
plt.plot(t, s, marker = 'o', label = 'posisi(m)')
plt.plot(t, v, marker = 's', label = 'kecepatan (m/s)')
plt.plot(t, a, marker = '^', label = 'percepatan(m/s)')
plt.title('gerak jatuh bebas')
plt.ylabel('nilai')
plt.grid(True)
plt.legend()
plt.show()
#nilai maksimum
print('hasil')
print(f'posisi maksimum: {max(s):.2f}')
print(f'kecepatan maksimum: {max(v):.2f}')
```
