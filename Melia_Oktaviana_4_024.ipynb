```python
#kasus 1
def simulasi_gerak_lurus():
    print('Simulasi Gerak Lurus dengan Percepatan Konstan')
    s0 = float(input('masukan posisi awal (m):'))
    v0 = float (input('masukan kecepatan awal (m/s):'))
    a = float (input('masukan percepatan (m/s²):'))
    t_total = int (input('masukan total Waktu simulasi (detik):'))
    dt = float (input('masukan interval Waktu (detik):'))

    t = 0
    print("nWaktu (s)\tPosisi (m)\tKecepatan (m/s)")
    while t <= t_total:
       s = s0 + v0 * t + 0.5 * a * t**2
       v = v0 + a * t
       print (f'{t:.1f}\t\t{s:.2f}\t\t{v:.2f}')
       t += dt

if __name__ == '__main__':
   simulasi_gerak_lurus()
```
